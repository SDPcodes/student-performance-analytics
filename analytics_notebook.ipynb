## Pre-Processing


#### 1. Check null value counts in each dataframe


```python
datasets = {
    "assessments": open_university_learning_analytics_assessments_ds,
    "courses": open_university_learning_analytics_courses_ds,
    "studentInfo": open_university_learning_analytics_student_info_ds,
    "studentRegistration": open_university_learning_analytics_student_registration_ds,
    "studentAssessment": open_university_learning_analytics_student_assessment_ds,
    "studentVle": open_university_learning_analytics_student_vle_ds,
    "vle": open_university_learning_analytics_vle_ds
}

for name, df in datasets.items():
    print(f"\n{name} missing values:")
    print(df.isnull().sum())
```

#### 2. Merge assessments and student assessment data sets


```python
open_university_learning_analytics_student_assessment_ds.dtypes
```

```python
open_university_learning_analytics_assessments_ds.dtypes
```

```python
# Change data type in student assessment dataset object -> numeric
open_university_learning_analytics_student_assessment_ds["score"] = pd.to_numeric(open_university_learning_analytics_student_assessment_ds["score"], errors="coerce").fillna(0)
open_university_learning_analytics_student_assessment_ds.dtypes
```

```python
# Merge Assessment Dataset into Student Assessment Dataset
open_university_learning_analytics_student_assessment_ds = open_university_learning_analytics_student_assessment_ds.merge(open_university_learning_analytics_assessments_ds, on="id_assessment", how="left")
open_university_learning_analytics_student_assessment_ds.head()
```

```python
# Create weighted score column
open_university_learning_analytics_student_assessment_ds["weighted_score"] = (open_university_learning_analytics_student_assessment_ds["score"] * open_university_learning_analytics_student_assessment_ds["weight"])/100
open_university_learning_analytics_student_assessment_ds.head()
```

```python
# Create an aggregated df which contains unique student id and their assessment mark summary
student_performance_aggregated_ds = open_university_learning_analytics_student_assessment_ds.groupby(["id_student"]).agg(
    total_weighted_score=('weighted_score', 'sum'),
    avg_score=('score', 'mean'),
    assessments_submitted=('id_assessment', 'count')
).reset_index()

student_performance_aggregated_ds.head()
```

#### 3. Merge engagement and student engagement data sets


```python
open_university_learning_analytics_student_vle_ds.dtypes
```

```python
open_university_learning_analytics_vle_ds.dtypes
```

```python
# Merge VLE dataset into Student VLE dataset
student_vle_merged = open_university_learning_analytics_student_vle_ds.merge(
    open_university_learning_analytics_vle_ds,
    on=["id_site", "code_module", "code_presentation"],
    how="left"
)

student_vle_merged.head()
```

```python
# Create an aggregated data frame which contains unique student id and their total clicks, unique activities, number of active days
student_vle_aggregated_ds = student_vle_merged.groupby(["id_student"]).agg(
    total_clicks = ("sum_click", "sum"),
    unique_activities = ("activity_type", "nunique"),
    active_days = ("date", "nunique"),
    total_vle_interactions = ("id_site", "count")
).reset_index()

student_vle_aggregated_ds.head()
```

#### 4. Compute Registration related features


```python
open_university_learning_analytics_student_registration_ds.dtypes
```

```python
# Convert data type of date_registration and date_unregistration to numerics
open_university_learning_analytics_student_registration_ds["date_registration"] = pd.to_numeric(open_university_learning_analytics_student_registration_ds["date_registration"], errors="coerce")
open_university_learning_analytics_student_registration_ds["date_unregistration"] = pd.to_numeric(open_university_learning_analytics_student_registration_ds["date_unregistration"].replace("?", pd.NA), errors="coerce")

open_university_learning_analytics_student_registration_ds.dtypes
```

```python
# Create an aggregated data frame which contains unique student id and their 
# earliest registration, whether they unregistered, unregistered module count, registered module count

student_registration_aggregated_ds = open_university_learning_analytics_student_registration_ds.groupby(["id_student"]).agg(
    earliest_registration=("date_registration", "min"),
    has_unregistered=("date_unregistration", lambda x: int(x.notna().any())),
    num_modules_registered=("code_module", "count"),
    num_modules_unregistered=("date_unregistration", "count")
).reset_index()

student_registration_aggregated_ds.head()
```

#### 5. Build the final merged master dataset


```python
# Start with Student Info dataset 
master_df = open_university_learning_analytics_student_info_ds.copy()

# Merge aggregated assessment ds
master_df = master_df.merge(student_performance_aggregated_ds, on="id_student", how="left")

# Merge aggregated VLE ds
master_df = master_df.merge(student_vle_aggregated_ds, on="id_student", how="left")

# Merge aggregated registration ds
master_df = master_df.merge(student_registration_aggregated_ds, on="id_student", how="left")

master_df.head()
```

#### 6. Handle missing values in dataset


```python
# Check null value counts in master dataset
master_df.isnull().sum()
```

```python
# Fill missing interactions, registrations and assessments with 0 which is make sense
numeric_fill_cols = ["total_weighted_score", "avg_score", "assessments_submitted",
                     "total_clicks", "unique_activities", "active_days",
                     "total_vle_interactions", "earliest_registration"]

for col in numeric_fill_cols:
    master_df[col] = master_df[col].fillna(0)
```

```python
# Overview of the final master dataset
print("=== Master Dataset Info ===")
print(f"Shape: {master_df.shape}")
print(f"\nData types:\n{master_df.dtypes}")
print(f"\n=== Statistical Summary ===")
master_df.describe()
```
