## Pre-Processing


#### 1. Check null value counts in each dataframe


```python
datasets = {
    "assessments": open_university_learning_analytics_assessments_ds,
    "courses": open_university_learning_analytics_courses_ds,
    "studentInfo": open_university_learning_analytics_student_info_ds,
    "studentRegistration": open_university_learning_analytics_student_registration_ds,
    "studentAssessment": open_university_learning_analytics_student_assessment_ds,
    "studentVle": open_university_learning_analytics_student_vle_ds,
    "vle": open_university_learning_analytics_vle_ds
}

for name, df in datasets.items():
    print(f"\n{name} missing values:")
    print(df.isnull().sum())
```

#### 2. Merge assessments and student assessment data sets


```python
open_university_learning_analytics_student_assessment_ds.dtypes
```

```python
open_university_learning_analytics_assessments_ds.dtypes
```

```python
# Change data type in student assessment dataset object -> numeric
open_university_learning_analytics_student_assessment_ds["score"] = pd.to_numeric(open_university_learning_analytics_student_assessment_ds["score"], errors="coerce").fillna(0)
open_university_learning_analytics_student_assessment_ds.dtypes
```

```python
# Merge Assessment Dataset into Student Assessment Dataset
open_university_learning_analytics_student_assessment_ds = open_university_learning_analytics_student_assessment_ds.merge(open_university_learning_analytics_assessments_ds, on="id_assessment", how="left")
open_university_learning_analytics_student_assessment_ds.head()
```

```python
# Create weighted score column
open_university_learning_analytics_student_assessment_ds["weighted_score"] = (open_university_learning_analytics_student_assessment_ds["score"] * open_university_learning_analytics_student_assessment_ds["weight"])/100
open_university_learning_analytics_student_assessment_ds.head()
```

```python
# Create an aggregated df which contains unique student id and their assessment mark summary
student_performance_aggregated_ds = open_university_learning_analytics_student_assessment_ds.groupby(["id_student"]).agg(
    total_weighted_score=('weighted_score', 'sum'),
    avg_score=('score', 'mean'),
    assessments_submitted=('id_assessment', 'count')
).reset_index()

student_performance_aggregated_ds.head()
```
